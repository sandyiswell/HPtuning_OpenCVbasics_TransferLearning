FILTERS

Filters generally use more than one pixel from the source image for computing each new
pixel value in the newly created image. 

Refer "Principles of Digital Image Processing:Fundamental Techniques" text book authored by Wilhelm Burger and Mark J. Burge for better theoretical understanding of the filtering concept. The following codes contain implementation of filtering from scratches using OpenCV in Google Colab .


```python
img=cv2.imread('/content/tgrcub.jpg')
cv2_imshow(img)
print(img.shape)
```

```python
## define the filter.
filter = np.array([
        [-1, -1, -1],
        [1, 1, 1],
        [0, 0, 0 ]]) # upper edge filter

# splitting the image for convolving the filter.
b,g,r=cv2.split(img)
channels=[b,g,r]
```

```python
channels = [b, g, r] # Three channels are made in to a list.
ch = 0

for channel in channels:
  for h in range(channel.shape[0]-2):   # since we are not considering padding, filtering needs to be done two rows & columns away.
    for w in range(channel.shape[1]-2):
      # Clipping the pixel value which can be accommodated. You can change this value if you want.
      max_pix_val = 145                  
      if channel[h +1, w +1] < max_pix_val:
          instance = channel[h:h+3, w:w+3]         # instance is nothing but the sliding window region. We are taking 3x3 region here.
          # h:h+3 defines three units of pixel dimensions across height.
          # w:w+3 defines three units of pixel dimensions across width.
          avg = int(np.sum(instance * filter)/ 9)  # average filtering. This is like a convolutional operation.
          channel[h +1, w +1] = avg                # new filtered intensity value is being assigned here.
  ch = ch +1                                       
  print(f"channel {ch} done.")       

merged = cv2.merge([b, g, r])
cv2_imshow(merged)
```
